```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
coefficients = [1,0,-13,-1,10,170]
roots=np.roots(coefficients)
a=roots.tolist()
s=0
for i in range(0,len(a)):
    if str(a[i])[-3] == '0' :
        s+=1
    else :
        continue
print('No.of real roots =',s)
print('\nThe roots are : ')
for i in range(len(a)):
    print(a[i])
print('\n')    

x=np.arange(-500,500,0.001)
sy=x*x*x*x*x -13*x*x*x -x*x +10*x +170
x=x.tolist()
y=sy.tolist()

f=plt.figure(1,figsize=(10,5))
#figure
plt.title('Graph of Polynomial')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.plot(x,y)
plt.grid()

#Greschgorin circle
figure, axes = plt.subplots(figsize=(8,5))
    
c1=plt.Circle((0,0),170, alpha = 0.5 , color='red')
c2=plt.Circle((0,0),10, alpha=0.5 , color='blue')
c3=plt.Circle((0,0),1 ,alpha=0.5 , color='green')
c4=plt.Circle((0,0),13 ,alpha=0.5 , color='yellow')
axes.add_artist(c1)
axes.add_artist(c2)
axes.add_artist(c3)
axes.add_artist(c4)
plt.text(20,400,'i')
axes.set_title('Gerschgorin Circles')


ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')





plt.xlim([-500,500])
plt.ylim([-500,500])

plt.show()
```

0 & 1 & 0 & 0 & 1 \\

0 & 0 & 0 & 1 & 0 \\
\end{bmatrix} 
$


# Greschgorin-Circle Theorem

$ Given\,a\,Companion\,matrix\,$$A{ij}$$\, the\,eigenvalues\,of\,the\,matrix\,lies\,in\,the\,union\,of\,discs\, $$ |z-A{ii}|\leq \sum\limits_{i \neq j} A{ij}\,.$

$ When\,a\,characteristic\,polynomial\,is\,converted\,to\,a\,companion\,matrix,\,the\,theorem\,states\,that\,the\,roots\,of\,the\,characteristic\,polynomial\,equal\,the\,eigenvalues\,of\,the\,companion\,matrix.\,Therefore,\,the\,roots\,of\,the\,polynomial\,lie\,in\,the\,union\,of\,all\,the\,discs\,.$
